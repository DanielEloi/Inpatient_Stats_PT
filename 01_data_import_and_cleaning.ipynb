```python
# The imports
import pandas as pd
```

```python
# Read the data file, rename columns, delete unnecessary data
data = pd.read_csv("data/raw/hospital_inpatient_activity.csv", sep=";")
data.columns = ["Month", "Region", "Institution", "Loc", "Specialty type", "Discharges", "Bed days"]
del data["Loc"]
```

```python
# Search if there are any duplicated values
dup_check = data[data.duplicated(subset=["Month", "Region", "Institution", "Specialty type"])]
if (dup_check.empty == True):
    print("Success - no duplicate entries of data for an institution's specialty type in the same month found.")
else:
    print("CAUTION: There seem to be some duplicate entires for at least one institution's specialty type in the same month.")
```

```python
# Fix the data types, set up the index 
data["Month"] = pd.to_datetime(data["Month"])
data.set_index("Month", inplace=True)
data = data.to_period("M")

data["Region"] = data["Region"].astype("category")
data["Institution"] = data["Institution"].astype("category")
data["Specialty type"] = data["Specialty type"].astype("category")
```

```python
# Save the dataFrame to a file
data.to_parquet("parsed_data.parquet")
```
